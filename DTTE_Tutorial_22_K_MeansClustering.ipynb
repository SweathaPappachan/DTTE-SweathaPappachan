# K-Means

K-Means algorithm is a centroid based clustering algorithm where the sum of distances of points from the centroid of each cluster is minimized. The final output is a set of K clusters .ie. the cluster assigned to each point and the K centroids of the clusters.

### The Algorithm
> 1.   Initialize K centroids to K points randomly and set each point's initial cluster as the centroid closest to it.
2.   While the clusers are changing - \\
a. Assign the new centroids as the centroids of the points which have the same assigned cluster. \\
b. Assign the new clusters to the points as the closest centroid.
3. Return the assignments and the centroids.





### Part 1 - Implementation of KMeans and testing on Synthetic Datasets

```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating Synthetic Data

k=2
color=["darkturquoise","darkorange","teal","darkviolet","tomato", "yellowgreen","hotpink","gold"]

X,_=make_blobs(n_samples=500, n_features=2, centers=k, random_state=10)
print("Shape = " + str(X.shape))

print("Sample:")
print(X[:7])
```

####Define all the functions & varibles needed to do clustering

```python
#Create an empty dictionary to store the cluster name, centroid and points associated with the cluster.
clusters={}

def initializeClusterCentroids():
  #For each cluster, initialize its properties
  for i in range(k):
      center = np.random.randint(-8, 10, size= (X.shape[1],), )
      points = []

      cluster = {
          'center' : center,
          'points' : points,
          'color' : color[i]
      }
      clusters[i] = cluster #looping over clusters dictionary

#Calculate euclidean distance between v1 and v2
def distance(v1, v2):
    return np.sum((v2-v1)**2)**0.5

def assignPointsToCluster(clusters, X):
    for ix in range(X.shape[0]):#iterate for all datapoints

        #Foe each datapoint, find the distance to k centroids
        distance_of_i = []
        for kx in range(k):
            d = distance(X[ix], clusters[kx]['center']) #we have find distance b/w
            #all the point with all the five (k) cluster centers
            distance_of_i.append(d)

        #Based on the distances to k centroid, pick which cluster to assign it to
        #here we want find the minimum distance of cluster centers
        cluster_to_choose = np.argmin(distance_of_i)
        clusters[cluster_to_choose]['points'].append(X[ix])

#STEP 3: UPDATE CLUSTER CENTROIDS
def updateCluster(clusters):
    for kx in range(k):
        cluster_points = clusters[kx]['points']

        cluster_points = np.array(cluster_points)
        #here pts are array of list but we want array of array so

        if len(cluster_points)>0:
            new_center = np.mean(cluster_points, axis=0) #axis is along rows so will
            #find mean of all the feature seprately it will give 2,0 when we have cluster pts 50,2
            clusters[kx]['center'] = new_center
            clusters[kx]['points'] = [] # clear the points in a cluster list (emptying pts)
            #as we have to do step 2 after step 3 again till not converge
```

```python
#We need to visualize this too. Lets define a function that can do that!
def plotclusters(clusters, plot):
    for kx in range(k):
        cluster_points = clusters[kx]['points']
        cluster_color = clusters[kx]['color']
        cluster_center = clusters[kx]['center']
        cluster_points = np.array(cluster_points)

        #plotting points associated /nearest to the cluster centers
        if len(cluster_points) > 0:
          plot.scatter(cluster_points[:, 0], cluster_points[:, 1], s = 2, c = cluster_color)

        plot.scatter(cluster_center[0], cluster_center[1], s = 250, c = cluster_color, marker="o")
```

We can now run and visualize how the clusters evolve.

```python
#Plotting the initial dataset
fig, axs = plt.subplots(2, 3)
fig.set_size_inches(15, 10)
axs[0, 0].scatter(X[:,0],X[:,1], s = 2, c = 'black')
axs[0, 0].set_title('Dataset')

np.random.seed(1)
initializeClusterCentroids()

#Now, lets plot the centroids it has initalized
axs[0, 1].set_title('Centroid Initialization')
axs[0, 1].scatter(X[:, 0], X[:, 1], s = 2, c = 'black')
for i in range(k):#for all the clusters
    center = clusters[i]['center'] #its cluster's center
    axs[0, 1].scatter(center[0], center[1], c=clusters[i]['color'], s=250, marker="o")

#CLUSTERING EPOCH 1
axs[0, 2].set_title('Iteration 1')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[0, 2])
updateCluster(clusters)

#CLUSTERING EPOCH 2
axs[1, 0].set_title('Iteration 2')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 0])
updateCluster(clusters)

#CLUSTERING EPOCH 3
axs[1, 1].set_title('Iteration 3')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 1])
updateCluster(clusters)

#CLUSTERING EPOCH 4
axs[1, 2].set_title('Iteration 4')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 2])
updateCluster(clusters)

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

### KMeans on Digits Dataset (KMeans for labelling)

Here we will attempt to use k-means to try to identify similar digits without using the original label information; this might be similar to a first step in extracting meaning from a new dataset about which you don't have any a priori label information.

```python
# importing the dataset
from sklearn.datasets import load_digits

digits = load_digits()
digits.data.shape
```

```python
# importing Kmeans from scikit-learn
from sklearn.cluster import KMeans

# Perform KMeans
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
# We can now see how the centers of the clusters formed are

fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

Because k-means knows nothing about the identity of the cluster, the 0–9 labels may be permuted. We can fix this by matching each learned cluster label with the true labels found in them:

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

Now we can check how accurate our unsupervised clustering was in finding similar digits within the data:

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

**Hence clustering algorithms like KMeans are highly effective for identifying labels for groups apriori and even a simple KMeans acheives a ~80% accuracy on digit classification.**

The confusion matrix for this looks like:

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

### KMeans on Iris Dataset (Elbow Method)

We will perform KMeans on the famous Iris Dataset which has four features: sepal length, sepal width, petal length, and petal width. The fifth column is for species, which holds the value for these types of plants.

```python
#Importing the neccesary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
#import the dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
x = df.iloc[:, [0,1,2,3]].values
df.head(5)
```

Let us start with assigning the initial number of clusters as k=5.

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

kmeans5.cluster_centers_
```

How do we choose the optimal number of clusters? There’s a method called the Elbow method, which is designed to help find the optimal number of clusters in a dataset. So let’s use this method to calculate the optimum value of k. To implement the Elbow method, we need to create some Python code (shown below), and we’ll plot a graph between the number of clusters and the corresponding error value.

This graph generally ends up shaped like an elbow, hence its name:

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)

plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

The output graph of the Elbow method shows that the shape of elbow is approximately formed at k=3. The optimal value of k is between 2 and 4, as the elbow-like shape is formed at k=3 in the above graph.

```python
kmeans3 = KMeans(n_clusters=3)
y_kmeans3 = kmeans3.fit_predict(x)
print(y_kmeans3)

kmeans3.cluster_centers_
```

We can now visualize the clustering in two dimensions instead of the original four.

```python
plt.scatter(x[:,0], x[:,1], c=y_kmeans3, cmap='rainbow')
```

The training complexity of KMeans is O(I * K * N * d), where:

- I is the number of iterations required for convergence (usually a small number, often less than 100).
- K is the number of clusters.
- N is the number of data points.
- d is the number of features (dimensionality of the data).

The testing complexity of KMeans, when you want to assign new data points to existing clusters, is O(K * d), where:

- K is the number of clusters.
- d is the number of features (dimensionality of the data).

KMeans typically converges to a solution in a small number of iterations, so the training complexity is often dominated by the number of data points (N) and the dimensionality of the data (d). The testing complexity is mainly influenced by the number of clusters (K) and the dimensionality of the data (d).

Changing the number of clusters (n_clusters) in K-Means can significantly affect the clustering outcome, and finding the optimal value for n_clusters is an essential step in K-Means clustering. The choice of the number of clusters should be based on the underlying structure of the data. Here's how changing the number of clusters can affect the outcome:

1. **Under-Clustering (Fewer Clusters)**:
   - If you choose a smaller value for n_clusters than the actual number of clusters in the data, K-Means may merge distinct clusters into one.
   - This can result in less granular and less accurate clustering, where clusters may not capture the inherent structure in the data.

2. **Over-Clustering (More Clusters)**:
   - If you choose a larger value for n_clusters than the actual number of clusters in the data, K-Means may split a single cluster into multiple smaller clusters.
   - This can lead to overfitting and create artificial subdivisions within clusters.

The Elbow Method is a common technique to find an optimal value for n_clusters. Here's how it works:

1. Run K-Means clustering with a range of values for n_clusters (e.g., from 1 to some upper limit) and record the sum of squared distances (inertia) for each clustering.

2. Plot the inertia values against the number of clusters.

3. Look for the "elbow point" on the plot, which is the point where the inertia starts to decrease at a slower rate. This point represents a good trade-off between model complexity and clustering quality.

4. The number of clusters at the elbow point is often considered as a reasonable choice for the optimal value of n_clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Feature matrix

# Define a range of values for n_clusters
n_clusters_range = range(1, 11)
inertia_values = []

# Calculate inertia (within-cluster sum of squares) for each value of n_clusters
for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Plot the inertia values to find the elbow point
plt.plot(n_clusters_range, inertia_values, marker='o')
plt.xlabel('Number of Clusters (n_clusters)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal n_clusters')
plt.grid(True)
plt.show()
```

In the resulting plot, you should look for the point where the inertia starts to decrease at a slower rate, resembling an "elbow." That point can be a good choice for the optimal value of n_clusters. Keep in mind that the optimal value may not always be perfectly clear-cut, and domain knowledge should also guide your selection.

The choice of initialization method in K-Means can indeed impact both the convergence speed and the quality of the clusters.
1. **K-Means++ Initialization**:
   - **Convergence Speed**: K-Means++ initialization typically converges faster than random initialization because it spreads the initial centroids apart, resulting in a better starting point for the clustering process.
   - **Quality of Clusters**: K-Means++ often leads to better quality clusters because it reduces the likelihood of converging to suboptimal solutions. It can result in more stable and accurate clusters.

2. **Random Initialization**:
   - **Convergence Speed**: Random initialization can be slower to converge because it starts with centroids placed randomly in the feature space. The initial placement may not be close to the true cluster centers.
   - **Quality of Clusters**: Random initialization is more likely to result in suboptimal solutions. The quality of clusters can vary, and there's a higher risk of getting stuck in local minima, leading to less accurate clustering.

The impact of initialization methods can be significant, and in practice, K-Means++ is often preferred over random initialization due to its better convergence properties and the likelihood of producing higher-quality clusters. However, it's essential to remember that K-Means can still be sensitive to the initial placement of centroids even with K-Means++.

To mitigate the risk of poor convergence or suboptimal solutions, a common practice is to run K-Means multiple times with different initializations (e.g., using K-Means++ with multiple restarts). This allows you to choose the best result based on a clustering evaluation metric, such as silhouette score or sum of squared distances, to achieve more robust and reliable clustering results.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn import metrics

# Load the Digits dataset
digits = datasets.load_digits()
X = digits.data  # Feature matrix
y = digits.target  # Ground truth labels

# Define a range of values for n_clusters
n_clusters_range = range(2, 11)
silhouette_scores = []
ari_scores = []

# Calculate silhouette scores and adjusted Rand indices for each value of n_clusters
for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    labels = kmeans.labels_

    # Calculate silhouette score
    silhouette_score = metrics.silhouette_score(X, labels, metric='euclidean')
    silhouette_scores.append(silhouette_score)

    # Calculate adjusted Rand index
    ari_score = metrics.adjusted_rand_score(y, labels)
    ari_scores.append(ari_score)

# Plot silhouette scores
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(n_clusters_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (n_clusters)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for K-Means Clustering')

# Plot adjusted Rand indices
plt.subplot(1, 2, 2)
plt.plot(n_clusters_range, ari_scores, marker='o')
plt.xlabel('Number of Clusters (n_clusters)')
plt.ylabel('Adjusted Rand Index')
plt.title('Adjusted Rand Index for K-Means Clustering')

plt.tight_layout()
plt.show()
```
